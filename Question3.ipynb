```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
# Load the dataset from CSV
data2 = pd.read_excel('/content/chatgpt1-2.xlsx')
```

```python
data = data2.head(1000)
```

```python
# Preprocessing
data = data.dropna(subset=['Text'])  # Remove rows with missing text content
text_data = data['Text'].values
```

```python
# Feature extraction using TF-IDF
vectorizer = TfidfVectorizer()
text_features = vectorizer.fit_transform(text_data)
```

```python
# Additional features
num_mentions = data['Text'].str.count('@')
num_hashtags = data['Text'].str.count('#')
other_features = np.column_stack((num_mentions, num_hashtags))
```

```python
# Combine text features with other features
combined_features = np.hstack((text_features.toarray(), other_features))
```

```python
# Perform clustering using K-Means
num_clusters = 5  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(combined_features)
```

```python
# Get the cluster labels
cluster_labels = kmeans.labels_
```

```python
# Add the cluster labels to the dataset
data['ClusterLabel'] = cluster_labels
```

```python
for index, row in data.iterrows():
    print(f"Tweet: {row['Text']}, Cluster: {row['ClusterLabel']}")
```
